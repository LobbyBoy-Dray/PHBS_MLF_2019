# Homework 1: playing with pandas dataframe

```python
from scipy.io import arff
import pandas as pd
```

```python
# Read data from .arff
data = arff.loadarff("4year.arff")
df   = pd.DataFrame(data[0])
df.head()
```

```python
# Add a column "bankruptcy"
df["bankruptcy"] = (df["class"]==b"1")
```

## 1. Create a new dataframe with only 4 feataures (and Bankruptcy). Properly rename the columns to X1, X2, X7, and X10

```python
# Select the specific columns and rename them
cols    = ["Attr1","Attr2","Attr7","Attr10","bankruptcy"]
col_map = {"Attr1":"r_profit",
           "Attr2":"r_liabilities",
           "Attr7":"r_EBIT",
           "Attr10":"r_equity"}
df_new = df[cols].rename(col_map, axis="columns")
df_new.head()
```

## 2. Fill-in the missing values na with the mean.

```python
# fill the missing value by mean
for col in df_new.columns:
    df_new[col].fillna(df_new[col].mean(), inplace=True)
```

```python
# check the missing values
df_new.info()
```

## 3. Find the mean and std of the 4 features among all, bankrupt and still-operating companies (3 groups).

```python
result = df_new.groupby("bankruptcy").agg(["mean","std"])
idx = pd.IndexSlice
result.loc["All", idx[:,"mean"]] = df_new.mean()[:-1].values
result.loc["All", idx[:,"std"]] = df_new.std()[:-1].values
result
```

## 4. How many companies have X1 values 1 std below the mean AND X10 values 1 std below the mean?

```python
mask1 = df_new["r_profit"] < (result.loc[idx["All"], idx["r_profit","mean"]] - result.loc[idx["All"], idx["r_profit","std"]])
mask2 = df_new["r_equity"] < (result.loc[idx["All"], idx["r_equity","mean"]] - result.loc[idx["All"], idx["r_equity","std"]])
df_new.loc[mask1 & mask2]
```

```python
print("【%s】 companies have X1 values 1 std below the mean AND X10 values 1 std below the mean." % len(df_new.loc[mask1 & mask2]))
```

## 5. What is the ratio of the bankrupted companies among the sub-groups above?

```python
(df_new.loc[mask1 & mask2]["bankruptcy"].sum())/15
```
